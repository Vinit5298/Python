```python
import random

# Start of the program
print("Welcome to the Banking System")

has_account = input("Do you have a bank account? (yes/no): ").lower()

# Store customer data
bank_data = {}
loan_data = {}

# If user doesn't have an account
if has_account == 'no':
    print("\nYou need to open a new bank account.")
    print("Choose account type:")
    account_types = [
        "Savings Account", "Current Account", "Fixed Deposit Account",
        "Recurring Deposit Account", "NRI (Non-Residential Indian) Accounts",
        "Senior Citizen Savings Scheme (SCSS)", "Joint Account", "Demat Account",
        "Salary Account", "Corporate Account", "Student Account", "Minorâ€™s Account"
    ]
    for i in range(len(account_types)):
        print(f"{i+1}. {account_types[i]}")
    acc_choice = int(input("Enter your choice (1-12): "))
    chosen_account = account_types[acc_choice - 1]

    print("\nPlease provide the following details:")
    name = input("1. Name: ")
    father = input("2. Father's Name: ")
    mother = input("3. Mother's Name: ")
    occupation = input("4. Occupation: ")
    address = input("5. Address: ")
    photo = input("6. Passport Size Photos (Y/N): ")
    aadhar = input("7. Aadhar Card (Y/N): ")
    pan = input("8. Pan Card (Y/N): ")
    mobile = input("9. Mobile Number: ")
    email = input("10. Email Address: ")

    # Generate unique 12-digit account number
    account_number = str(random.randint(100000000000, 999999999999))
    print("\nYour account has been successfully created!")
    print("Your Account Number is:", account_number)

    bank_data['name'] = name
    bank_data['account_number'] = account_number
    bank_data['account_type'] = chosen_account
    bank_data['balance'] = 0
    bank_data['mobile'] = mobile
    loan_data['loan_balance'] = 0
    loan_data['emi_months_left'] = 0

else:
    # Assume the user has an existing account
    print("\nAccessing your account...\n")
    # Just creating dummy data for existing account for demo
    bank_data['name'] = "Vinit"
    bank_data['account_number'] = "123456789012"
    bank_data['account_type'] = "Savings Account"
    bank_data['balance'] = 100000
    loan_data['loan_balance'] = 0
    loan_data['emi_months_left'] = 0

    while True:
        print("\nChoose Option:")
        print("1. Account Services")
        print("2. Loan Services")
        print("3. Exit")

        service_choice = int(input("Enter your choice: "))

        if service_choice == 1:
            print("\nAccount Services:")
            print("1. Withdrawal Amount")
            print("2. Deposit Money")
            print("3. Check Balance")
            print("4. Transfer Money to another account")
            acc_option = int(input("Enter option: "))

            if acc_option == 1:
                amount = int(input("Enter amount to withdraw: "))
                if amount <= bank_data['balance']:
                    bank_data['balance'] -= amount
                    print("Withdrawal successful. Remaining Balance:", bank_data['balance'])
                else:
                    print("Insufficient Balance.")

            elif acc_option == 2:
                amount = int(input("Enter amount to deposit: "))
                bank_data['balance'] += amount
                print("Deposit successful. Total Balance:", bank_data['balance'])

            elif acc_option == 3:
                print("Your Current Balance is:", bank_data['balance'])

            elif acc_option == 4:
                name = input("Enter recipient name: ")
                acc_no = input("Enter recipient account number: ")
                amount = int(input("Enter amount to transfer: "))
                if amount <= bank_data['balance']:
                    bank_data['balance'] -= amount
                    print(f"Transferred {amount} to {name} ({acc_no}). Remaining Balance:", bank_data['balance'])
                else:
                    print("Insufficient Balance.")

        elif service_choice == 2:
            print("\nLoan Services")
            amount = int(input("How much loan do you need? "))
            print("Checking eligibility...")

            # Assume eligibility if the loan is <= 3x the balance
            if amount <= bank_data['balance'] * 3:
                print("Eligible for Loan. Loan will be credited with 10% interest.")
                loan_data['loan_balance'] = int(amount * 1.1)  # 10% interest
                loan_data['emi_months_left'] = 6
                emi = loan_data['loan_balance'] // 6
                bank_data['balance'] += amount
                print(f"Loan of {amount} credited to your account.")
                print(f"EMI of {emi} will be deducted for next 6 months.")
                print("Updated Balance:", bank_data['balance'])
            else:
                print("Sorry, you are not eligible for this loan amount.")

        elif service_choice == 3:
            print("Thank you for using our services.")
            break

# Summary
print("\n--- Customer Summary ---")
for key, value in bank_data.items():
    print(f"{key.capitalize()}: {value}")
if loan_data['loan_balance'] > 0:
    print("Loan Balance:", loan_data['loan_balance'])
    print("Remaining EMIs:", loan_data['emi_months_left'])
```
