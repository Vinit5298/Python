```python
# 4. Write a program to print all even numbers between 1 to 100. – using while loop.

i = 1
n = 100
sum = 0

while i <= n:
    if i % 2 == 0:
        sum = sum + i
        print(i, end=" ")
    i = i + 1

print("\nSum of All Even Numbers from 1 to 100 is:", sum)
```

```python
# 14.Write a program to find LCM of two numbers using while loop.

n = int(input("Enter first number: "))
m = int(input("Enter second number: "))

# Start from the greater of the two numbers
lcm = max(n, m)

while True:
    if lcm % n == 0 and lcm % m == 0:
        print(f"The LCM of {n} and {m} is: {lcm}")
        break
    lcm = lcm + 1
```

```python
# 16.Write a program to print all Prime numbers between 1 to n using while loop.

n = int(input("Enter a Number: "))
num = 2  # Start from 2, since 1 is not a prime

while num <= n:
    count = 0
    i = 1
    while i <= num:
        if num % i == 0:
            count = count + 1
        i = i + 1
    if count == 2:
        print(f"{num} is a Prime Number")
    num = num + 1
```
